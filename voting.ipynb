```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import  seaborn as sns
from scipy.stats import uniform as sp_randFloat
from scipy.stats import randint as sp_randInt 
# %matplotlib


# Importing the dataset
dataset = pd.read_csv('dataset/hackathon.csv/Train.csv')
# dataset.describe()
# dataset.info()


c_l = pd.get_dummies(dataset['form_field47'],drop_first=True)
d_s = pd.get_dummies(dataset['default_status'],drop_first=True)

dataset = pd.concat([dataset,c_l,d_s],axis=1)
dataset.drop(['form_field47','default_status','Applicant_ID'],axis=1,inplace=True)
```

```python
dataset
```

```python
dataset.fillna(-999,inplace=True)
```

```python
from sklearn.preprocessing import MinMaxScaler
```

```python
scalar = MinMaxScaler()
```

```python
dataset = pd.DataFrame(scalar.fit_transform(dataset),
                      columns=dataset.columns)
dataset
```

```python
X = dataset.iloc[:,0:50]

y = dataset.iloc[:,-1]
```

```python

from sklearn.model_selection import train_test_split
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.2, 
                                                    random_state=0)
```

```python
from sklearn.ensemble import RandomForestClassifier
from catboost import CatBoostClassifier
from lightgbm import LGBMClassifier
from sklearn.ensemble import VotingClassifier
```

```python
rf = RandomForestClassifier(n_estimators=500)
lg = LGBMClassifier()
cb = CatBoostClassifier(depth=5,learning_rate=0.035,n_estimators=700)
cb1 = CatBoostClassifier()
```

```python
rf.fit(X_train,y_train)
lg.fit(X_train,y_train)
cb.fit(X_train,y_train)
cb1.fit(X_train,y_train)
```

```python
classifiers = [('rf',rf),('lg',lg),('cb',cb),('cb1',cb1)]
```

```python
vc = VotingClassifier(estimators=classifiers,voting='soft',
                     n_jobs=-1,verbose=1)
```

```python
vc.fit(X_train,y_train)
```

```python
y_pred = vc.predict_proba(X_test)[:,1]
```

```python


from sklearn.metrics import confusion_matrix,classification_report,roc_auc_score,auc
# cm = confusion_matrix(y_test,y_pred) 
# print(cm)
# cp = classification_report(y_test,y_pred)
# print(cp)
auc = roc_auc_score(y_test,y_pred)
print(auc)
# fpr, tpr, thresholds = roc_curve (y_test, y_pred)
# roc_auc= auc (fpr, tpr)
# print  ("ROC AUC", roc_auc)
```
